```python
def correlation(X, Y):
    X = X - np.mean(X, axis=0)
    Y = Y - np.mean(Y, axis=0)
    return np.sum(X * Y, axis=0) / (np.sqrt(np.sum(np.square(X), axis=0)) * np.sqrt(np.sum(np.square(Y), axis=0)))

def test(trace, plain):
    KeyCorrs  = []
    MaxPeak   = []
    Ratio     = []

    corrs    = []
    maxCorrs = []
    for GuessKey in tqdm(range(256)):
        IntermediateValue = HW[np.array(Sbox)[(plain[:] ^ GuessKey).astype(int)]]
        corr = np.abs(correlation(trace, IntermediateValue))
        corrs.append(corr)
        maxCorrs.append(np.max(corr))
    key        = np.argmax(maxCorrs)
    RankedCorr = np.sort(maxCorrs)
    
    KeyCorrs.append(corrs)
    MaxPeak.append(maxCorrs)
    Ratio.append(RankedCorr[-1] / RankedCorr[-2]) 
        
    return key, Ratio, KeyCorrs, MaxPeak
```

```python
def get_batches(data, batch_size):
    batches = []
    for i in range(int(data.shape[0] // batch_size)):
        batch = data[i * batch_size: (i + 1) * batch_size]
        batches.append(batch)
    return np.asarray(batches)
```

```python
class Stretch_Sigmoid(layers.Activation):
    def __init__(self, activation, **kwargs):
        super(Stretch_Sigmoid, self).__init__(activation, **kwargs)
        self.__name__ = 'stretch_sigmoid'

def stretch_sigmoid(x, beta=0.1):
    return K.sigmoid(beta * x)

keras.utils.get_custom_objects().update({'stretch_sigmoid': Stretch_Sigmoid(stretch_sigmoid)})
```

```python
def Generator(points_len):
    input_layer = layers.Input(shape = (points_len, ))
    
    Dense_1 = layers.Dense(5, activation = 'relu')(input_layer)
    
    output_layer = layers.Dense(points_len, activation = 'sigmoid', name="Gen_layer")(Dense_1)
    
    return tf.keras.Model(inputs=[input_layer], outputs=[output_layer])
```

```python
def Predictor(points_len):
    input_layer = layers.Input(shape = (points_len, ))
    
    Dense_1 = layers.Dense(30, activation = 'relu')(input_layer)
    
    output_layer = layers.Dense(8, activation = 'stretch_sigmoid', name="Predict_layer")(Dense_1)
    
    return tf.keras.Model(inputs=[input_layer], outputs=[output_layer])
```

```python
def Test_DNN(points_len):
    input_layer = layers.Input(shape = (points_len, ))
    trace_layer = layers.Input(shape = (points_len, ))
    
    emm = layers.Embedding(input_dim = points_len, output_dim = points_len)(input_layer)
    lstm = layers.LSTM(10)(emm)
    
    gen_noise = layers.Dense(points_len, activation = 'tanh', name="Gen_layer")(lstm)
    gen_noise_batch = layers.BatchNormalization()(gen_noise)
    
    gen_trace = layers.multiply([gen_noise_batch, trace_layer])
    
    Dense_2 = layers.Dense(30, activation = 'relu')(gen_trace)
    
    output_layer = layers.Dense(1, activation = 'stretch_sigmoid', name="Predict_layer")(Dense_2)
    
    return tf.keras.Model(inputs=[input_layer, trace_layer], outputs=[output_layer])
```

```python
def build_SelfAattention_with_MLP(points_len):
    input_layer = layers.Input(shape = (points_len, ))
    #bat_input = layers.BatchNormalization()(input_layer)
    #attention_probs_1 = layers.Dense(points_len, activation='softmax')(input_layer)
    #attention_mul_1 = layers.multiply([input_layer, attention_probs_1])
    Dense_1 = layers.Dense(50, activation = 'relu')(input_layer)
    
    attention_probs_2 = layers.Dense(50, activation='softmax')(Dense_1)
    attention_mul_2 = layers.multiply([Dense_1, attention_probs_2])
    Dense_2 = layers.Dense(50, activation = 'relu')(attention_mul_2)
    y = layers.Dense(1, activation = 'sigmoid')(Dense_2)
    
    return tf.keras.Model(inputs=[input_layer], outputs=y)
```
